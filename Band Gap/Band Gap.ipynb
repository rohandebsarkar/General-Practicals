```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import scienceplots
plt.style.use(['science','ieee', 'grid'])
plt.rcParams["figure.figsize"] = (10, 6)

Z = 5e-4
```

```python
def st_line_fit(x1, y1):
    S = len(x1)
    S_xy = np.sum(x1*y1)
    S_x = np.sum(x1)
    S_y = np.sum(y1)
    S_xx = np.sum(x1*x1)
    D = S*S_xx-S_x**2

    m1 = (S*S_xy - S_x*S_y) / D
    c1 = (S_xx*S_y - S_x*S_xy) / D

    return m1, c1
```

```python
data1 = pd.read_csv("data1.csv")

data1
```

```python
x1=data1["1/T"][10:]
y1=data1["ln Rho"][10:]

m1, c1 = st_line_fit(x1, y1) 
print(m1)

plt.plot(x1, y1,".")
plt.plot(x1, x1*m1 + c1)
plt.show()

```
