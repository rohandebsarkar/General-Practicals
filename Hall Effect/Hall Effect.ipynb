```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import scienceplots
plt.style.use(['science','ieee', 'grid'])
plt.rcParams["figure.figsize"] = (10, 6)

Z = 5e-4
```

## Constant B

```python
data1 = pd.read_csv("data1.csv")

data1
```

```python
B = np.unique(data1["B (KG)"])
x=np.zeros((4, 5))
y=np.zeros((4, 5))
m=np.zeros(4)
for i in range(len(B)):
    x[i] = data1[data1["B (KG)"] == B[i]]["I (mA)"]
    y[i] = data1[data1["B (KG)"] == B[i]]["Mean Hall Voltage (mV)"]
    plt.plot(x[i], y[i], ".")

for i in range(len(x)):
    m[i] = np.sum(x[i] * y[i]) / np.sum(x[i]**2)
    plt.plot(np.append(0, x[i]), m[i]*np.append(0, x[i]), label=f"B = {B[i]}")

plt.legend()
plt.show()
```

```python
R=np.zeros(4)
for i in range(len(m)):
    R[i]=m[i]*Z/B[i]

np.mean(R)
```
